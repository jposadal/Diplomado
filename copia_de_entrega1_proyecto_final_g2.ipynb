```python
casos = pd.read_csv('/content/Diplomado/CASOS.csv', sep=';', dtype=str)
casos.info()
casos.head()
```

```python
casos['codigo_municipio'] = casos['codigo_municipio'].astype("string")
casos.info()
casos.head()
```

```python
casos['cantidad'] = casos['cantidad'].astype("int64")
casos['delito'] = casos['delito'].astype("int64")
casos['anno'] = casos['anno'].astype("int64")
casos.info()
casos.head()
```

```python
delitos = pd.read_csv('/content/Diplomado/DELITOS.csv', sep=';', dtype=str)
delitos.info()
delitos.head()
```

```python
delitos['Codigo Delito'] = delitos['Codigo Delito'].astype("int64")
delitos.info()
delitos.head()
```

```python
municipios = pd.read_csv('/content/Diplomado/Municipios.csv', sep=';', dtype=str)
municipios.info()
municipios.head()
```

```python
municipios = municipios.drop ('Unnamed: 3', axis = 1)
municipios.head ()
```

```python
municipios['codigo_municipio'] = municipios['codigo_municipio'].astype("string")
municipios['codigo_departamento'] = municipios['codigo_departamento'].astype("string")
municipios.head ()
municipios.info()
```

```python
municipios.head()
```

```python
departamentos = pd.read_csv('/content/Diplomado/departamentos.csv', sep=';', dtype=str)
departamentos.info()
departamentos.head()
```

```python
departamentos['codigo_departamento'] = departamentos['codigo_departamento'].astype("string")
departamentos.info()
departamentos.head()
```

```python
gini = pd.read_csv('/content/Diplomado/gini.csv', sep=';', dtype=str)
gini.info()
gini.head()
```

```python
gini['codigo departamento'] = gini['codigo departamento'].astype("string")
gini['anno'] = gini['anno'].astype("int64")
gini['cantidad'] = gini['cantidad'].str.replace(',', '.')
gini['cantidad'] = gini['cantidad'].astype("float64")
gini.info()
gini.head()
```

```python
poblacion = pd.read_csv('/content/Diplomado/poblacion.csv', sep=';', dtype=str)
poblacion.info()
poblacion.head()
```

```python
poblacion['codigo_municipio'] = poblacion['codigo_municipio'].astype("string")
poblacion['codigo_departamento'] = poblacion['codigo_departamento'].astype("string")
poblacion['anno'] = poblacion['anno'].astype("int64")
poblacion ['poblacion'] = poblacion['poblacion'].str.replace('.', '')
poblacion['poblacion'] = poblacion['poblacion'].astype("float64")
poblacion.info()
poblacion.head()
```

```python
merged_df = pd.merge(casos, poblacion, on=['codigo_municipio', 'anno'], how='inner')
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df["indice"] =round(((merged_df["cantidad"]/merged_df["poblacion"])*100000), 2)
print(merged_df)
```

```python
merged_df_dep = pd.merge(casos, municipios, on=['codigo_municipio'], how='inner')
merged_df_dep.head()
```

```python
merged_df_dep.info()
```

```python
merged_df_dep = merged_df_dep.drop ('municipio', axis = 1)
merged_df_dep = merged_df_dep.drop ('codigo_municipio', axis = 1)

```

```python
merged_df_dep.info()
```

```python
merged_df_dep['delito'] = merged_df_dep['delito'].astype("str")
merged_df_dep['anno'] = merged_df_dep['anno'].astype("str")
merged_df_dep['codigo'] = merged_df_dep.anno.str.cat([merged_df_dep.codigo_departamento, merged_df_dep.delito], sep="_")
```

```python
merged_df_dep.info()
merged_df_dep.head()
```

```python
merged_df_dep['delito'] = merged_df_dep['delito'].astype("int64")
merged_df_dep['anno'] = merged_df_dep['anno'].astype("int64")
merged_df_dep.info()
```

```python
merged_df_dep_group = merged_df_dep.groupby('codigo').sum('cantidad')
```

```python
merged_df_dep_group = merged_df_dep.groupby('codigo')['cantidad'].sum().reset_index()
```

```python
resultado = pd.DataFrame(merged_df_dep_group)
```

```python
resultado
```

```python
depagrupado = pd.merge(resultado, merged_df_dep, on=['codigo'], how='left')
```

```python
depagrupado
```

```python
depagrupado.info()
depagrupado.head()
```

```python
depagrupado.columns = ['codigo', 'cantidad', 'anno', 'delito', 'codigo_departamento']
depagrupado.info()
```
